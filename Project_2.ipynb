## What is a Dipole?

A dipole refers to the distribution of electrical charges within a molecule, resulting in to one end containing a partial positive charge (δ⁺) and the other end containing a partial negative charge (δ⁻). Dipoles occur in molecules that have polar bonds where one atom is more electronegative than the other and that possess the right geometry to ensure the bond dipoles do not cancel each other.

As an example carbon dioxide does not have a dipole moment. It does contain polar bonds, but its linear geometry causes these dipole to cancel each other. However, Ammonia is a great example of a molecule that does have a dipole moment. Ammonia contains polar bonds and due it is pyramidal shape these dipoles do not cancel each other whichleads to an net dipole upward toward the nitrogen atom. A visual of ammonia's dipole is shown below.

## What is a Quadrupole?
Quadrupole describes the electric charge configuration across a non-polar molecules that contains regions of partial positive and negative charges due to its specific bond polarities and geometry. Quadrupole occurs when there are two pairs of positive and negative charges arranged with alternating positive and negative regions. Pattern one is partial positive, partial negative, partial negative, partial positive. Pattern two is partial negative, partial postive, partial positive, partial negative.

An example of a molecule possessing a quadrupole is Carbon Dioxide (CO2). As the oxygens are more electronegative causing them to posses a partial negative charge. The Carbon is less electronegative cause it to possess a particle positive chagre. The linear natural of carbon dioxide and the presence of two carbon oxygen bonds results in the molecules having partial negative, partial postive, partial positive, partial negative which allows it to be a quadropole.

## Molecules Containing Both a Dipole and Quadrupole?

 Molecules that contain two or more dipoles that are arranged asymmetrically, contain both a dipole and quadrupole. The molecule will contain a net dipole as the sum of the individual bond dipoles will not cancel each other. The molecule will also have a quadrupole moment as the molecule’s charge distribution forms an alternating positive and negative potential which go beyond the dipole.

 A common example of such a molecule is water. Water contains two polar bonds, which due to its geometry, are arrangment asymmetrically. Therefore the dipole moments of the two H-O bonds do not cancel each other allowing the molecule to possess a net dipole moment towards the oxygen. It however, still contains the alternating partial positive partial charge, double negative particle charge, partial positive partial charge pattern resulting in it also containing a quadrupole moment.

## Particle in a Box in an External Electric Field

Without an external field the energies and wavefunction is as followed:
$$
E = \frac{n^2 h^2}{8ma^2}
$$

$$
\psi_n(x) = \sqrt{\frac{2}{a}}sin{\frac{n\pi x}{a}}
$$

**The perturbed energy will be:**

When a uniformed electric fieldis ($\mathcal{E}$) is applied along the box, the potential energy term is $V(x)=−q\mathcal{E} x$ where q is the charge of the particle. The new Hamiltonian becomes $\hat{H} = \hat{H_0}−q\mathcal{E} x$  

$$
E_{n}^0 = \left\langle \psi_n | -q\mathcal{E} x |\psi_n \right\rangle\
$$

Given the unperturbed provided wavefunctions above, this intergal will give:

$$
\left\langle \psi_n |x|\psi_n \right\rangle\ ={\frac{a}{2}}
$$

making the first order energy shift:

$$
E_{n}^1 =-q\mathcal{E}{\frac{a}{2}}
$$

There is an uniform shift of all levels by
$-q\mathcal{E}{\frac{a}{2}}$
 resulting in no change in spacing between energy levels.

**The perturbed wavelength will be:**

The wavefunction is no longer symmetric. The field tilts the potential, resulting in one side of the well being lower in energy and one side being higher in energy. The wavefunction will shift towards the lower potential meaning that is more likely to find the particle in the side with lower potential.

## Thinking about Dipole and Quadrupole Specifically
For our specific system we are going to think of potential energy in the form of a dipole and quadrupole moment. The potential is given as:

$V(x) = F\cdot x + G\cdot x^2$

*   $F\cdot x$ represents the interaction of the particle with a uniform external field. The linear dependence on x shows that this term couples to the dipole moment of the particle. The dipole potential is given as $V(x)_{dipole} = -q\mathcal{E} x$. This term tilts the potential resulting in the energy increasing or decreasing linearly with x. This asymmetric potenital will shift the wavefunction towards the side of lower potential. The tilt of the potential depends on the direction of the electric field. Thinking speciically about an electron (negative charge $q=-e$)

if $$F > 0$$

The potential will increase with x causing the electron to pull to the left


if $$F < 0$$

The potential will decrease with x causing the electron to pull to the right

The potential forms a ramp with the slant depending on the field direction. The probability of finding the electron will always be higher in the area of lower potential. When $F > 0$ the slope of the ramp is positive meaning that the left side is lower potential making it more likely to find the electron of the left side of the box. When $F < 0$ the slope of the ramp is negative meaning that the right is lower in potential making it more likely to find the electron of the right side of the box.

*   $G\cdot x^2$ represent the system's interaction with a non-uniform external field. The quadratic dependence on x shows that this term couples to the quadrupole moment of the particle. This term shifts the box into a parabola shape comparable to the shape of a harmonic oscillator. Instead of pulling the electron with the left or right of the box, this term shifts the electron to either towards the centre of the box or away from the centre of the box.

if $$G > 0$$

The potential will form a parabolic shape, with $x=0$ being the minimum and the potnetial increasing towards the walls.

if $$G < 0$$

The potential will form an inverted parabolic shape, with $x=0$ being the maximum and the potnetial decreasing towards the walls.

The probability of finding the electron will always be higher in the area of lower potential. When $G > 0$ the potential minimum occurs at x=o meaning that it is most likely to find the electron in the center of the box. When $G < 0$ the potentium minimum occurs at the walls of the box meaning that is in most likely to find the electron at the walls of the box.

### **Polynomial Basis**

Chebyshev polynomials are chosen as the basis functions in this Jupyter notebook for several important reasons:

*   Chebyshev polynomials $ T_k(x) $ form an **orthogonal** basis on the interval $ -1 \leq x \leq 1 $
*   **Minimize numerical errors** and avoid issues like Runge's phenomenon where large oscillations occur near the edges of the interval, leading to poor approximations
*   Provide an **optimal way to approximate smooth functions** with exponential convergence in many cases
*   Allows for fast evaluation and differentiation using recurrence relations, making them **computationally efficient**


The polynomial class implemented in NumPy can be used to generate the basis functions. This defines polynomials, by default, on the interval
$-1 < x < 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial
from scipy.integrate import quad
from scipy.linalg import eigh
from scipy.optimize import minimize
from numpy.polynomial import Chebyshev as T

def get_basis(n):
    """
    Return the N+1 basis functions for k = 0, 1, ..., n. The basis functions
    are defined in the default interval -1 < x < 1 and have the form:

        phi_k(x) = (1-x)(1+x)T_k(x)

    Parameters
    ----------
    n: int
        n+1 is the number of basis functions. n must be a positive integer.

    Returns
    -------
    list of Chebyshev polynomials representing the basis functions.
    """
    if n < 1:
        raise ValueError("n must be at least 1")

    # Define polynomials (1-x) and (1+x) in Chebyshev form
    p1 = Polynomial([-1, 1]).convert(kind=T)
    p2 = Polynomial([1, 1]).convert(kind=T)

    # Generate basis functions using Chebyshev polynomials
    phi = [(p1 * p2 * T.basis(k)).convert(kind=T) for k in range(n+1)]

    return phi
```

### **About the Potential**
For the given potential,
$$
V(x) = F\cdot x + G\cdot x^2 + H\cdot x^3
$$
*   $F$ is the force due to the applied electric field interacting with a dipole moment.
*   $G$ is associated with an interaction that depends on the field strength squared and relates to quadrupole effects.
*   $H$ is associated with octupole effects, representing higher-order nonlinear interactions that introduce asymmetry in the potential.


Therefore, for the three terms:

*   $F\cdot x$ is a linear potential, which represents the interaction of a dipole moment with an external electric field.

*   $G\cdot x^2$ is a quadratic potential that is related to the interaction involving quadratic polarizability, which describes how the system's response to an external field depends on the square of the field strength.

*   $H\cdot x^3$ is a cubic potential linked to octupole effects, introducing asymmetry in the potential due to higher-order field interactions. This term represents the interaction of an external field with an intrinsic octupole moment or field-induced dipole-quadrupole coupling, contributing to nonlinear energy shifts.

```python
from numpy.polynomial.chebyshev import Chebyshev as T
from numpy.polynomial.polynomial import Polynomial

# For now, the H parameter is set to zero, but can be changed
def potential(F, G, H = 0):
    """
    Define the potential as a polynomial object, including contributions
    from dipole and quadrupole polarizability.

    The potential is defined as:

        V(x) = F * x + G * x^2 + H * x^3

    Parameters
    ----------
    F : float
        Coefficient related to the potential from dipole polarizability.
    G : float
        Coefficient related to the potential from quadrupole polarizability.
    H : float
        Coefficient related to the potential from octupole polarizability.

    Returns
    -------
    Chebyshev polynomial representation of the potential.
    """
    # The potential is kept as Chebyshev polynomial to avoid polynomial type error
    return F * T([0, 1]) + G * T([0, 0, 1]) + H * T([0, 0, 0, 1])
```

### **Potential Matrix**
Elements in the diagonal of the matrix is:
$$
v_{jk} = \int_{-1}^{1} \phi_j(x) V(x) \phi_k(x) \, dx
$$

This is calculated in the function below.

```python
def compute_v(phi, F, G, H = 0):
    """
    Returns the potential energy matrix for the basis polynomials phi,
    using the potential V(x) = F*x + G*x^2 + H*x^3.

    Parameters
    ----------
    phi : list of Polynomial
        Array of basis functions.
    F : float
        Coefficient related to the potential from dipole polarizability.
    G : float
        Coefficient related to the potential from quadrupole polarizability.
    H : float
        Coefficient related to the potential higher-order nonlinear interactions.

    Returns
    -------
    numpy.ndarray
        The potential energy matrix.
    """
    from numpy.polynomial import Polynomial

    n = len(phi)
    v = np.zeros((n, n))

    # Define the potential function (ensured to be a Chebyshev potential)
    V_x = potential(F, G, H=0)

    # Compute the matrix elements
    for j in range(n):
        for k in range(n):
            # Form the product polynomial
            phij_v_phik = phi[j] * V_x * phi[k]
            # Compute the definite integral from -1 to 1
            integral = phij_v_phik.integ()
            v[j, k] = integral(1.0) - integral(-1.0)

    return v
```

### **Kinetic Energy Matrix**
Note: this is in atomic units. \
Elements in the diagonal of the matrix is:
$$
t_{jk} = -\frac{1}{2} \int_{-1}^{1} \phi_j(x) \frac{d^2}{dx^2} \phi_k(x) \, dx
$$



```python
def compute_t(phi):
    """
    Returns the kinetic energy matrix for the basis polynomials phi

    Parameters
    ----------
    phi: Polynomial array of basis functions
    """

    # Start by filling the kinetic energy matrix with zeros.
    n = len(phi)
    t = np.zeros((n, n))

    # Populate the kinetic energy matrix
    for j in range(n):
        for k in range(n):
            # Take the second derivative of the second basis function and
            # multiply it by the first basis function, then multiply by -1/2
            phij_d2_phik = -1./2 * phi[j] * phi[k].deriv(2)
            # Do the *indefinite* integral
            integral = phij_d2_phik.integ()
            # Evaluate the definite integral by taking the difference of the
            # polynomial evaluated at two points.
            t[j,k] = integral(1.0) - integral(-1.0)
    return t
```

### **Overlap Matrix**
Elements in the diagonal of the matrix is:
$$
s_{jk} = \int_{-1}^{1} \phi_j(x) \phi_k(x) \, dx
$$

These are calculated in the functions below.

```python
def compute_s(phi):
    """
    Returns the overlap matrix for the basis polynomials phi

    Parameters
    ----------
    phi: Polynomial array of basis functions
    """

    # Start by filling the overlap matrix with zeros.
    n = len(phi)
    s = np.zeros((n, n))

    # Populate the overlap matrix
    for j in range(n):
        for k in range(n):
    ### START YOUR CODE HERE
            # Form the product polynomial
            phij_phik = phi[j]*phi[k]
            # Do the *indefinite* integral
            integral = phij_phik.integ()
            # Evaluate the definite integral by taking the difference of the
            # polynomial evaluated at two points.
            s[j,k] = integral(1.0) - integral(-1.0)
    ### END YOUR CODE HERE
    return s
```

### **Secular Equation**

The secular equation is given by

$$
H c = E S c
$$

where,
- $H$ is the Hamiltonian matrix with elements, $h_{jk} = t_{jk} + v_{jk}$
- $ S $ is the overlap matrix with elements, $s_{jk}$
- $ c $ is the coefficient vector,
- $ E $ represents the eigenvalues (energies).

The function below solves the generalized eigenvalue problem:
$$
\sum_{k=0}^{\infty} h_{jk} c_k = E \sum_{k=0}^{\infty} s_{jk} c_k
$$

```python
def solve_secular_eq(phi,F,G,H=0):
    """
    Returns the eigenvalues and eigenvectors of the secular equation, with
    a basis of polynomials specified by phi and a perturbation strength f.

    Parameters
    ----------
    phi: Polynomial array of basis functions
    f: strength of perturbation
    """

    # Start by filling the potential matrix with zeros.
    n = len(phi)
    s = np.zeros((n, n))

    ### START YOUR CODE HERE:
    # Compute the overlap matrix, kinetic energy matrix, and potential energy matrix
    s = compute_s(phi)
    t = compute_t(phi)
    v = compute_v(phi, F, G, H=0)

    # Compute the Hamiltonian matrix.
    h = t + v

    # Solve the generalized eigenvalue problem using np.linalg.eigh
    # return the eigenvectors and eigenvalues of eigenvectors and eigenvalues as c and E, respectively,
    return eigh(h, s)
    ### END YOUR CODE HERE
```

### **Wavefunction**
We can construct the **trial wavefunction** as a linear combination of basis functions $ \phi_k(x) $ with coefficients $ c_k $:

$$
\psi(x) = \sum_{k=0}^{N} c_k \phi_k(x)
$$

#### Parameters:
- `c`: A vector containing the wavefunction expansion coefficients.
- `phi`: A vector of polynomial basis functions.

#### Returns:
- A polynomial object representing $ \psi(x) $.


```python
def get_wavefunction(phi,c):
    """
    Return a Polynomial object representing the trial wavefunction expanded
    in the basis defined by phi with coefficients c,

         \psi(x) = \sum_k=0^N c_k \phi_k(x)

    Parameters:
    ----------
    c :: a vector of N+1 wavefunction coefficients.
    phi :: a vector of polynomial basis functions.
    """
    return sum([c[k]*phi[k] for k in range(len(c))])
```


### **Eigenenergies**

We can compute the **eigenenergies** for a given basis set size $N$ and perturbation strengths $ F $, $ G $, and optionally $ H $.

#### Steps:
1. The basis functions are generated using `get_basis(n)`.
2. The secular equation is solved using `solve_secular_eq(phi, F, G, H)`, which returns the eigenvalues and eigenvectors.
3. The eigenvalues, which correspond to the system's energy levels, are returned.


#### Parameters:
- `n`: The size of the basis set.
- `F`: The strength of the applied perturbation (e.g., external electric field effect).
- `G`: The strength of the quadratic perturbation (e.g., quadrupole effects).
- `H`: (Optional) A parameter representing higher-order perturbation effects.

#### Returns:
- A list of eigenvalues (energy levels of the system).

```python
def get_energies(n,F,G,H=0):
    """
    A convenient function just to return the eigenenergies for a given
    basis set size, N, and perturbation strength F.

         \psi(x) = \sum_k=0^N c_k \phi_k(x)
    """
    phi = get_basis(n)
    eigenvalues, eigenvectors = solve_secular_eq(phi,F,G, H=0)

    return eigenvalues
```

### **Variational Principle**

A method used to approximate the ground state energy by minimizing a functional over the set of trial wavefunctions. It gives a lower bound for the true ground state energy.

$$
\min_{\kappa} \langle \Psi(\kappa) | \hat{H} | \Psi(\kappa) \rangle
$$


```python
def variational_principle(phi, F, G, H=0):
    """
    Approximate the ground state energy using the variational principle.
    Parameters
    ----------
    phi : list
        Basis functions.
    F, G, H : float
        Potential parameters.
    Returns
    -------
    float
        Approximated ground state energy.
    """

    n = len(phi)

    # Compute matrices
    s = compute_s(phi)
    t = compute_t(phi)
    v = compute_v(phi, F, G, H)

    h = t + v

    # Trial coefficients are normalized
    def energy(c):
        c = np.array(c)
        return (c @ h @ c) / (c @ s @ c)

    # Initial guess
    c0 = np.random.rand(n)

    # Normalize initial guess
    c0 /= np.sqrt(c0 @ s @ c0)

    # Minimize the energy functional
    result = minimize(energy, c0, method='BFGS')

    E_var = energy(result.x)
    return E_var
```

### **Perturbation Theory**

A method to calculate the correction to the ground state energy due to a small perturbation, starting with the unperturbed energy and applying first-order corrections based on the perturbation matrix.

#### 1. Computing First-Order Energy Correction (*E₁*)

The correction is computed as:

$$
E_1 = \frac{\langle \psi_0 | \hat{V} | \psi_0 \rangle}{\langle \psi_0 | \psi_0 \rangle}
$$

#### 2. Computing First-Order Wavefunction Correction (*ψ₁*)

This solves:

$$
(\hat{H}_0 - E_0) | \psi_1 \rangle = (\hat{V} - E_1) | \psi_0 \rangle
$$



```python
import numpy as np

def perturbation_theory(phi, F0, G0, H0=0, dF=0, dG=0, dH=0):
    """
    Compute first-order perturbation correction to ground state energy and wavefunction.

    Parameters
    ----------
    phi : list
        Basis functions.
    F0, G0, H0 : float
        Unperturbed potential parameters.
    dF, dG, dH : float
        Perturbation strength (small values).

    Returns
    -------
    (float, float, np.ndarray)
        E0 : Unperturbed ground state energy
        E1 : First-order correction
        psi1 : First-order correction to the wavefunction
    """

    # Unperturbed problem
    E, c = solve_secular_eq(phi, F0, G0, H0)
    c0 = c[:, 0]  # ground state

    # Overlap matrix
    s = compute_s(phi)

    # Perturbation matrix
    V1 = compute_v(phi, dF, dG, dH)

    # Compute unperturbed Hamiltonian matrix
    t = compute_t(phi)  # Compute kinetic energy matrix
    v = compute_v(phi, F0, G0, H0)  # Compute potential energy matrix
    H0_matrix = t + v  # Hamiltonian matrix

    # First-order energy correction
    E1 = (c0.T @ V1 @ c0) / (c0.T @ s @ c0)

    # Compute first-order wavefunction correction
    psi1 = np.linalg.solve(H0_matrix - E[0] * np.eye(len(phi)), (V1 - E1 * np.eye(len(phi))) @ c0)

    # Return E0, E1, and first-order wavefunction correction
    return E[0], E1
```
